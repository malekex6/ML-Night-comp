```python
def find_boundaries(df,variable):
    Q1 = df[variable].quantile(.25)
    Q3 = df[variable].quantile(.75)
    IQR=Q3-Q1
    lower_boundary=Q1-2*IQR
    upper_boundary=Q3+2*IQR
    return lower_boundary, upper_boundary
```

```python
data.shape
```

```python
lower_days,upper_days=find_boundaries(data,"days")
outliers_days=np.where(data["days"]>upper_days,True ,np.where(data["days"]<lower_days,True,False))
data = data.loc[~(outliers_days),]

lower_surface,upper_surface=find_boundaries(data,"surface")
outliers_surface=np.where(data["surface"]>upper_surface,True ,np.where(data["surface"]<lower_surface,True,False))
data = data.loc[~(outliers_surface),]
```

```python
y=data['budget']
x=data.drop(['budget'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1, shuffle=True)
```

```python
from xgboost import XGBRegressor
from sklearn.ensemble import AdaBoostRegressor,BaggingRegressor,GradientBoostingRegressor,RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
models=[AdaBoostRegressor(),BaggingRegressor(),GradientBoostingRegressor(),RandomForestRegressor(),LinearRegression(),XGBRegressor()]
for model in models:
    model.fit(X_train,y_train)
    y_pred=model.predict(X_test)
    err=mean_squared_error(y_test,y_pred,squared=True)
    print("model",model)
    print("err",err)
    print("score",model.score(X_test,y_test))
    print("#######")
    
```
